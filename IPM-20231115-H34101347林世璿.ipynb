$$
\begin{array}{rl}
\max & 4y_1+4y_2\\

\end{array}

The KKT condition of
$$
\begin{array}{rl}
\min & -x_1-3x_2\\
s.t. & x_1-2x_2 \geq -4\\
     & -x_1-x_2\geq -4\\
     &x_1,x_2\geq 0
\end{array}
$$

\begin{array}{l}
(P)\left\{
\begin{array}{l}
\color{red}{x_1-2x_2-x_3=-4}\\
\color{red}{-x_1-x_2-x_4= -4}\\
x_1,x_2,x_3,x_4\geq 0\\
\end{array}
\right.\\
(D)\left\{
\begin{array}{l}
\color{red}{y_1-y_2+y_3 =-1}\\
\color{red}{-2y_1-y_2+y_4 =-3}\\
y_1,y_2,y_3,y_4\geq 0\\
\end{array}
\right.\\

\begin{array}{l}

\end{array}

\end{array}
$$
