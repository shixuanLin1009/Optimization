```python
import numpy as np
def E1(A,i,k):
    m=len(A)
    n=len(A[0,:])
    I=np.identity(m)
    I[i,i]=k
    return I.dot(A)
def E2(A,i,j,k):
    m=len(A)
    n=len(A[0,:])
    I=np.identity(m)
    I[j,i]=k
    return I.dot(A)
```

```python
class LPOption:
    def __init__(self):
        self.Maximize = 0
        self.Minimize = 0
        self.ConstraintType=[]
        self.VariableUnresticted=[]
```

```python
def solveLP(init_A,init_B,init_C,optimalType,constrainType,initialSolution):
    
    M = 20000
    
    if initialSolution != True:  #還沒有初始解
        #lt eq gt is constrain type
        lt,eq,gt = 0,0,0
        for i in constrainType:
            if i == 0:
                lt = lt + 1
            elif i == 1:
                eq = eq + 1
            elif i == 2:
                gt = gt + 1
        
        m = 1 + len(init_A)
        n = 1 + len(init_A[0]) + 1 + 1 + lt + eq + gt * 2
        np.set_printoptions(precision=3,suppress=True)
        #print(lt,eq,gt)
        #print(m,n)
        A = np.zeros((m,n))
        
        
    
        
        if optimalType == 1:   #max problem
            A[0,0] = 1
            for i in range(len(init_C)):
                A[0,1+i] = -init_C[i][0]
        else:  #min problem
            A[0,0] = -1
            for i in range(len(init_C)):
                A[0,1+i] = init_C[i][0]
            
        for i in range(len(init_A)):
            for j in range(len(init_A[0])):
                A[i+1,j+1] = init_A[i][j]
        for i in range(len(init_B)):
            A[i+1,-2] = init_B[i][0]
            
            
        #set basic variable BV
        t = 1 + len(init_A[0]) 
        for i in range(len(init_A)):
            if constrainType[i] == 0:
                A[i+1,t] = 1
                
                t = t + 1
            elif constrainType[i] == 1:
                A[i+1,t] = 1
                A[0,t] = M
                A = E2(A,i+1,0,-M)
                t = t + 1
            elif constrainType[i] == 2:
                A[i+1,t] = -1
                A[i+1,t+1] = 1
                A[0,t+1] = M
                A = E2(A,i+1,0,-M)
                t = t + 2
        print("Initial Tabuleau :")
        
        print(A)
    else:  #已經有初始解
        lt,eq,gt = 0,0,0
        for i in constrainType:
            if i == 0:
                lt = lt + 1
            elif i == 1:
                eq = eq + 1
            elif i == 2:
                gt = gt + 1
        
        m = 1 + len(init_A)
        n = 1 + len(init_A[0]) + 1 + 1 
        #print(lt,eq,gt)
        #print(m,n)
        A = np.zeros((m,n))
        if optimalType == 1:   #max problem
            A[0,0] = 1
            for i in range(len(init_C)):
                A[0,1+i] = -init_C[i][0]
        else:  #min problem
            A[0,0] = -1
            for i in range(len(init_C)):
                A[0,1+i] = init_C[i][0]
            
        for i in range(len(init_A)):
            for j in range(len(init_A[0])):
                A[i+1,j+1] = init_A[i][j]
        for i in range(len(init_B)):
            A[i+1,-2] = init_B[i][0]
            
            
        print("Initial Tabuleau :")
        print(A)

    num = 1
    
    rq=-1
    while rq<0: 
        print("########Iteration " + str(num) +"########")
        num += 1
        #######################################
        rq=float("inf")
        j=-1
        for l in range(1,n-2):
            if A[0,l]<rq:
                rq=A[0,l]
                j=l
        #print(rq,j)
        if rq == 0:
            print("The end tabuleau!!")
            break
        else:    
            print("Choose " + str(rq) + ", at column " + str(j))
        #######################################    
        if rq>=0:
            break
        #######################################
        MRT=float("inf")
        i=-1
        for l in range(1,m):
            if A[l,j]>0:
                A[l,n-1]=A[l,n-2]/A[l,j]
                if A[l,n-2]/A[l,j]<MRT:
                    MRT=A[l,n-2]/A[l,j]
                    i=l
        #print(MRT,i)
        print("min = " + str(MRT) + ", at row " + str(i))
        print(A)
        #######################################
        A=E1(A,i,1/A[i,j])
        A=E2(A,i,0,-A[0,j])
        for l in range(1,i):
            A=E2(A,i,l,-A[l,j])
        #for l in range(0,i):    
        #    A=E2(A,i,0,-A[l,j])
        for l in range(1,i):
            A=E2(A,i,l,-A[l,j])
        for l in range(i+1,m):
            A=E2(A,i,l,-A[l,j])
        #######################################
        for i in range(m):
            A[i,n-1]=0

        #print(A)

    print(A)  
    if optimalType == 1:
        print("Optimal value : " + str(A[0,-2]))
    else:
        print("Optimal value : " + str(-A[0,-2]))
    
```
